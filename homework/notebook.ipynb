```python
import pandas as pd  #  type: ignore
import zipfile
import os
```

```python
# Ruta al archivo .zip
zip_file_path = '../files/input.zip'

# Carpeta a donde se extraen los archivos
input_folder = '../input'

# Carpeta donde se guardaran los archivos finales
ouput_folder = '../files/output'
```

```python

# Verificar si la carpeta existe
os.makedirs(input_folder, exist_ok=True)
os.makedirs(ouput_folder, exist_ok=True)
```

```python
# Extraer el archivo .zip

with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(input_folder)
```

```python
    
# Para leer los archivos en la carpeta test
data_test = []
categories = ['negative', 'positive', 'neutral']


for category in categories:
    category_path = os.path.join('../input/input/test', category)
    if not os.path.exists(category_path):
        raise FileNotFoundError(f"The directory {category_path} does not exist.")
    
    for filename in os.listdir(category_path):
        if filename.endswith(".txt"):
            file_path = os.path.join(category_path, filename)
            with open(file_path, 'r', encoding='utf-8') as file:
                content = file.read()
                data_test.append({'phrase': content, 'target': category})
                
```

```python
dataframe = pd.DataFrame(data_test)
```

```python
 #Save the DataFrame to a CSV file if needed
dataframe.to_csv('../files/output/test_dataset.csv', index=False)
```
